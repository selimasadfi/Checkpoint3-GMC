```python
#Libraries used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1-Load the dataset into a data frame using Python
data=pd.read_csv("Video games.csv")
data
```

```python
data.shape
```

```python
#2-Clean the data as needed
#i'm going to check it first and see what i need to change , fill or add
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
#i'm going to fill missing data from Year
data['Year'].fillna(data['Year'].median(),inplace=True)
data.isnull().sum()
```

```python
#now for the publisher
data['Publisher'].value_counts()
```

```python
data['Publisher'].fillna(data['Year'].median(),inplace=True)

data.isnull().sum() #we got rid of the missing values
```

```python
data
```

```python
#we're going to remove the rank cuz we don't need it
data.drop('Rank',axis=1, inplace=True)
```

```python
data
```

```python
#i'm going to also remove the Name column as it's not really mentionned in any of the questions 
data.drop('Name',axis=1, inplace=True)
```

```python
data
```

```python
#3-Create an area chart to show the trend of global sales over the years.
plt.figure(figsize=(12, 6))
plt.fill_between(data['Year'], data['Global_Sales'], alpha=0.9)
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.title('Global Sales Trend Over the Years')
plt.show()
```

```python
#4-Create a histogram to show the distribution of global sales.
plt.figure(figsize=(10, 6))
plt.hist(data['Global_Sales'], bins=20 , range=(0,20))


plt.xlabel('Global Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Global Sales (0 to 20 range)')
plt.grid(True)
plt.show() 
```

```python
#5-Create a scatter plot to show the relationship between North American sales and European sales.
plt.scatter(data['NA_Sales'], data['EU_Sales'])

plt.xlabel('North American Sales')
plt.ylabel('European Sales')
plt.title('Relationship Between North American Sales and European Sales')
```

```python
#6-Create a scatter plot to show the relationship between Japanese sales and Other sales.
plt.scatter(data['JP_Sales'], data['Other_Sales'])

plt.xlabel('Japan Sales')
plt.ylabel('Other_Sales')
plt.title('Relationship between Japanese sales and Other sales')
```

```python
#7-Create a stacked area chart to show the distribution of sales by platform over the years.

plt.stackplot(data.index, data['NA_Sales'], data['EU_Sales'], data['JP_Sales'],data['Other_Sales'],labels=['NA', 'EU', 'JP','O'])
plt.xlim(0, 7000) 
plt.ylim(0, 10)

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Distribution of Sales by Platform Over the Years')
```

```python
#8-Create a bar chart to show the top 10 publishers by global sales
top_publishers = data.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
top_publishers
```

```python
plt.figure(figsize=(12, 6))
top_publishers.plot(kind='bar')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (in millions)')
plt.grid(axis='y')
plt.show()
```

```python
#9-Create a pie chart to show the distribution of genres by global sales.
```

```python
data['Genre'].unique()
```

```python
genre_sales = data.groupby('Genre')['Global_Sales'].sum()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(genre_sales, labels=genre_sales.index,autopct='%1.0f%%',  startangle=140) #autopct='%1.0f%%' to show the percentage
plt.title('Distribution of Genres by Global Sales')
```
